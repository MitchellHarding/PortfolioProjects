```python
#WELCOME TO PYTHON MOVIE DATABASE CORRELATION PORTFOLIO

#importing packages etc
###############################
import pandas as pd
import numpy as np
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#putting data in 

#using dataframe 
df = pd.read_csv(r'C:\Users\Mitchell\Downloads\movies.csv')
```

```python
#the data
df.head()
```

```python
#checking for missing data

for col in df.columns:
 pct_missing = np.mean(df[col].isnull())
 print('{} - {}%'.format(col, pct_missing))
```

```python
#what are the data types

df.dtypes
```

```python
#get rid of decimal on budget

df['budget'] = df['budget'].astype('int64')
```

```python
#get rid of decimal on gross sales
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#creating correct year column
df['YearCorrect'] = df['released'].astype(str).str[:4]

df
```

```python
df.sort_values (by=['gross'], inplace =False, ascending = False)
```

```python
#displays whole data set
pd.set_option('display.max_rows', None)
```

```python
#eliminate any duplicate values, using .head to reduce scale size of output for viewing
df.head()['company'].drop_duplicates().sort_values(ascending = False)

#to eliminate duplicates from whole data set can use this
df.drop_duplicates()

```

```python
df.head()
```

```python
###Analysing correlations between different variables

#company vs gross profit

#budget vs gross profit

#year vs gross profit

#genre vs budget
```

```python
#buliding scatterplot
#budget vs gross revenue

plt.scatter (x=df['budget'], y = df['gross'])

plt.title('Budget Vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget For Film')

plt.show
```

```python
#regression plot using seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws= {"color": "black"}, line_kws={"color":"red"})
```

```python
#find the direct correlation stats

df.corr(method = 'pearson') #pearson is default, kendall, spearman
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix with Numeric Labels')

plt.xlabel('Movie Attributes')

plt.ylabel('Movie Attributes')

plt.show()
```

```python
#correlation between company and gross profit

#company is not numeric, cretaing the column as a numeric values

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
       df_numerized[col_name] = df_numerized[col_name].astype ('category')
       df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix with Numeric Labels')

plt.xlabel('Movie Attributes')

plt.ylabel('Movie Attributes')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs [(sorted_pairs) > 0.5]
high_corr
```
